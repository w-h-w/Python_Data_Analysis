# 电商数据评估清理步骤


**整体流程**
- 读取数据
- 对数据进行评估
- 通过评估数据得到的结论去解决一个个问题
- 每解决一个问题都回过头查看问题是否解决
- 最后把清洗后的数据保存为新的csv文件

## 了解表格基本信息 
**比如**
- 分析目标\
    此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收，本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。
- 简介\
    原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。
- 数据每列的含义如下：
1. InvoiceNo：发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母"c"开头，表示这笔交易被取消。
2. StockCode：产品代码。5位数，作为产品的唯一标识符。·Description：产品名称。
3. Quantity：产品在交易中的数量。
4. InvoiceDate：发票日期和时间。交易发生的期和时间。 UnitPrice：单价。价格单位为英镑（E）。
5. CustomerID：客户编号。5位数，作为客户的唯一标识符。 Country：国家名称。客户所居住的国家的名称。

## 读取数据

```python
import pandas as pd
```

**dtype = str,最好读取的时候都以字符串的形式读入，不然可能会使数据失真**\
**比如一个0010008的编号可能会读取成10008**

```python
original_data = pd.read_csv("data.csv", encoding="latin-1",dtype = str)
original_data.head()
```

## 评估数据
在这一部分，我将对在上一部分建立的original_data这个DataFrame所包含的数据进行评估。\
评估主要从两个方面进行：**结构和内容，即整齐度和干净度**。数据的结构性问题指不符合“**每列是一个变量，每行是一个观察值，每个单元格是一个值**” \
这三个标准**，数据的内容性问题包括存在丢失数据、重复数据、无效数据等。

### 评估数据整齐度


- 随机抽出表格中的行.sample()
- 观察是否符合--**每列是一个变量，每行是一个观察值，每个单元格是一个值**

```python
original_data.sample(10)
```

### 评估数据干净度

- info()方法可以知道数据的变量个数，变量数据类型，非空数量等dataframe的一些基本信息

```python
original_data.info()
```

- 从输出结果来看，数据共有541909条观察值，而Description、CustomerID变量存在缺失值。
- 此外，InvoiceDate的数据类型应为日期，
- CustomerID的数据类型应为字符串，应当进行数据格式转换。

#### 评估缺失数据

- 提取缺失值所在列通过isnull方法，Description中存在缺失值返回ture

```python
original_data["Description"].isnull() 
```

- 在了解Description存在缺失值后，根据条件提取出缺失观察值。筛选为ture的行

```python
original_data[original_data["Description"].isnull() ]
```

```python
original_data[(original_data["Description"].isnull()) & (original_data['UnitPrice'].astype('float') > 0)]
```

- 筛选出来结果数量为o条，说明缺失Description值的数据，同时也不具备有效的UnitPrice值。
- Description表示产品名称，UnitPrice表示产品单价，都是进行后续商品交易分析的重要变量。如果它们同时缺失/无效，我们认为数据无法提供有效含义，因此这些后续可以被删除。

```python
original_data[original_data["CustomerID"].isnull()]
```

- CustomerID表示客户编号，不是分析畅销商品的必要变量。并且从输出结果来看，有些CustomerID缺失的交易数据仍然有效，因此保留此变量为空的观察值。
  

#### 评估重复数据

```python
original_data["UnitPrice"] = original_data["UnitPrice"].astype("float")
```

```python
original_data["Quantity"] = original_data["Quantity"].astype("float")
```

#### 评估不一致数据

- 不一致数据可能存在于Country变量中，我们要查看是否存在多个不同值指代同一国家的情况。

```python
original_data["Country"].value_counts()
```

#### 评估无效或者错误数据

```python
original_data.describe()
```

**先筛选出Quantity数值为负数的观察值**

```python
original_data[original_data["Quantity"] < 0 ]
```

```python
original_data[(original_data["Quantity"] < 0) & (original_data["InvoiceNo"].str[0] != "C")]
```

```python
original_data[(original_data["Quantity"] < 0) 
            & (original_data["InvoiceNo"].str[0] != "C") 
            & (original_data["UnitPrice"] != 0)]
```

根据输出结果，猜想得到验证，当Quantity变量为负数的时候，观察值满足以下条件之一：
- InvoiceNo以C开头，表示订单被取消
- UnitPrice为0，说明单价为0英镑\

这些交易数据均不是有效成交数据，不贡献销售，不在后续分析范围内，因此我们将在数据清理步骤，将Quantity为负数的观察值删除。

**接下来筛选出UnitPrice为负数的观察值**

```python
original_data[original_data["UnitPrice"] < 0]
```

## 清理数据

根据前面评估部分得到的结论，我们需要进行的数据清理包括：
- InvoiceDate变量的数据类型转换为为日期时间
- 把CustomerID~变量的数据类型转换为字符串 Description变量缺失的观察值删除
- 把Country变量值“USA替换为United States”
- 把Country变量值"UK”，"U.K."替换为"United Kingdom”
- 把Quantity变量值为负数的观察值删除
- 把UnitPrice变量值为负数的观察值删除

```python
cleaned_data = original_data.copy()
cleaned_data
```

```python
cleaned_data["InvoiceDate"] = pd.to_datetime(cleaned_data["InvoiceDate"])
cleaned_data["InvoiceDate"]
```

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].astype(str)
cleaned_data["CustomerID"] 
# 在pandas中，python字符串表示为object，所以这个sh位置转换成功
```

```python
cleaned_data.dropna(subset=["Description"], inplace=True)
cleaned_data
```

```python
cleaned_data["Description"].isnull().sum()
```

```python
cleaned_data["Country"] = cleaned_data["Country"].replace({"USA":"United States"})
```

```python
len(cleaned_data[cleaned_data["Country"]=="USA"])
```

```python
cleaned_data["Country"] = cleaned_data["Country"].replace({"UK":"United Kingdom","U.K.":"United Kingdom"})
print(len(cleaned_data[cleaned_data["Country"]=="UK"]))
print(len(cleaned_data[cleaned_data["Country"]=="U.K."]))
```

```python
cleaned_data = cleaned_data[cleaned_data["Quantity"] > 0]
len(cleaned_data[cleaned_data["Quantity"] < 0])
```

```python
cleaned_data = cleaned_data[cleaned_data["UnitPrice"] >= 0]
len(cleaned_data[cleaned_data["UnitPrice"] < 0])
```

## 保存清理后的数据
完成数据清理后，把干净整齐的数据保存到新的文件里，文件名为data_cleaned.csv。

```python
cleaned_data
```

```python
cleaned_data.to_csv("data_cleaned.csv",index = False)
```
